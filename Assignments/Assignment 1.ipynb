```python
sentence = input("enter the sentence : ")
words = sentence.lower().split()
word_count = {}
for word in words :
    if word in word_count :
        word_count[word] += 1
    else :
        word_count[word] = 1
for word, count in word_count.items():
    print(word, ":", count)        

```

```python
def is_prime(n) :
    if n<2: 
        return False
    for i in range(2,n):
        if n % i == 0 :
            return False
    return True    
nums =   [2, 4, 5, 10, 13, 17, 20, 23]
prime_nums = []
for num in nums :
    if is_prime(num) :
        prime_nums.append(num)
print(prime_nums)  
```

```python
import pandas as pd
data = {
    "Product" : ["Laptop","Phone","Tablet","Monitor","Keyboard","Mouse","Printer","Camera"],
    "Price"   : [800.0, 500.0, 300.0, 200.0, 50.0, 25.0, 150.0, 600.0],
    "Quantity": [5, 10, 7, 8, 20, 30, 4, 6]
}
df = pd.DataFrame(data)
df["TotalValue"] = df["Price"] * df["Quantity"]
df_sorted = df.sort_values("TotalValue",ascending=False) 
grouped = df.groupby("Price")["Quantity"].mean()
print("Original DataFrame:\n", df)
print("\nSorted by TotalValue (descending):\n", df_sorted)
print("\nAverage Quantity per Price group:\n", grouped)
```

```python
import pandas as pd
df = pd.read_csv("train.csv")
print("\nMissing values:")
print(df.isnull().sum())
under_18 = df[df["Age"]<18]
print("\nPassengers under 18 years old:")
print(under_18[["Name", "Age", "Sex", "Survived"]])
female_survivors = df[(df["Sex"]=="female")&(df["Survived"]==1)]
print("\nFemale passengers who survived:")
print(female_survivors[["Name", "Sex", "Survived"]])
average_fare_by_class = df.groupby("Pclass")["Fare"].mean()
print("\nAverage Fare by Passenger Class:")
print(average_fare_by_class)
```

```python
import pandas as pd
import numpy as np
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
column_names = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS',
                'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope',
                'NumVesselsFluoro', 'Thalassemia', 'HeartDisease']
df = pd.read_csv(url, names=column_names)
df.replace('?', np.nan, inplace=True)
numeric_columns = ['NumVesselsFluoro', 'Thalassemia', 'Cholesterol']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
print("Missing values per column:\n")
print(df.isnull().sum())
df['NumVesselsFluoro'] = df['NumVesselsFluoro'].fillna(df['NumVesselsFluoro'].median())
df['Thalassemia'] = df['Thalassemia'].fillna(df['Thalassemia'].mode()[0])
df = df[df.isnull().sum(axis=1) <= 2]
df['HeartDisease'] = df['HeartDisease'].apply(lambda x: 1 if int(x) > 0 else 0)
cholesterol_avg = df.groupby('HeartDisease')['Cholesterol'].mean()
print("\nAverage Cholesterol by Heart Disease status:")
print(cholesterol_avg)
```

```python
import matplotlib.pyplot as plt
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
math_scores = [85, 78, 92, 88, 76]
science_scores = [90, 82, 89, 94, 75]
english_scores = [78, 85, 88, 80, 82]
history_scores = [84, 80, 91, 79, 77]
plt.figure(figsize=(8, 5))
plt.bar(names, math_scores, color='skyblue')
plt.title("Math Scores of Students")
plt.xlabel("Students")
plt.ylabel("Math Score")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
plt.figure(figsize=(10, 6))
plt.plot(names, math_scores, marker='o', label='Math')
plt.plot(names, science_scores, marker='s', label='Science')
plt.plot(names, english_scores, marker='^', label='English')
plt.plot(names, history_scores, marker='d', label='History')
plt.title("Student Performance in All Subjects")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mean = 50
std_dev = 10
data = np.random.normal(mean, std_dev, 1000)
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='lightgreen', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean}")
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
